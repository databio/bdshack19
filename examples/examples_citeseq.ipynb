```python
# Load the counts matrix
counts_file = homedir+'citeseq/processed_data/human_sub.csv.gz'
counts = pd.read_csv(counts_file,sep=',',index_col=0)
# rows are gene IDs, columns are cell barcodes
```

```python
counts.head(10)
```

```python
# determine sample labels
labels_file = homedir+'citeseq/processed_data/CBMC_ADT_clr_filtered.csv'
labels = pd.read_csv(labels_file,sep=',',index_col=0)
# rows are cell surface markers, columns are cell barcodes
```

```python
# merge the sample labels and counts to toss into loom
alldata = pd.concat([counts,labels])
alldata
```

```python
df_row_metadata = {'measurement_id':np.asarray(alldata.index.tolist())}
df_col_metadata = {'sample_id':np.asarray(alldata.columns.tolist())}
data = alldata.values

filename = 'testloompy.loom'
loompy.create(filename, data, df_row_metadata, df_col_metadata)
```
